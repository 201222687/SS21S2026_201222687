```python

# ==============================================
# TAREA 1 - LIMPIEZA Y PREPARACIÓN DE DATOS
# Dataset: student_productivity_distraction_dataset_20000
# Autor: Juan Carlos Maldonado Solórzano
# ==============================================


#IMPORTACION
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
# CARGA DEL DATASET

df_original = pd.read_csv("student_productivity_distraction_dataset_20000.csv")

print("Primeras filas del dataset original:")
df_original.head()
```

```python
# INFORMACION GENERAL

print("===== INFORMACIÓN GENERAL =====")
df_original.info()

print("\n===== ESTADÍSTICAS DESCRIPTIVAS =====")
df_original.describe()
```

```python
# VERIFICACION DE DUPLICADOS

print("Registros originales:", len(df_original))

duplicados = df_original.duplicated().sum()
print("Duplicados encontrados:", duplicados)
```

```python
# ELIMINACION DE DUPLICADOS (PROCEDIMIENTO FORMAL)

df = df_original.drop_duplicates()

print("Registros después del proceso:", len(df))
```

```python
# REVISION DE VALORES NULOS

print("Valores nulos por columna:")
df.isnull().sum()
```

```python
# TRATAMIENTO FORMAL DE NULOS
# Identificar columnas numéricas
columnas_numericas = df.select_dtypes(include=np.number).columns

# Aplicar imputación por media (aunque no existan nulos)
for col in columnas_numericas:
    df[col] = df[col].fillna(df[col].mean())

print("Verificación final de nulos:")
df.isnull().sum().sum()
```

```python
# ESTANDARIZACION

# Estandarizar género
df["gender"] = df["gender"].str.strip().str.upper()

print("Valores únicos en gender:")
df["gender"].unique()
```

```python
# DATASET LIMPIO

print("Dataset después del proceso de limpieza:")
df.head()
```

```python
# TABLA PIVOTE PRODUCTIVIDAD POR GENERO
tabla_genero = pd.pivot_table(
    df,
    values="productivity_score",
    index="gender",
    aggfunc="mean"
)

tabla_genero
```

```python
# TABLA PIVOTE PRODUCTIVIDAD SEGUN HORAS DE ESTUDIO
tabla_estudio = pd.pivot_table(
    df,
    values="productivity_score",
    index="study_hours_per_day",
    aggfunc="mean"
)

tabla_estudio.head()
```

```python
# CORRELACION (ANALISIS MAS PROFESIONAL)

correlaciones = df.corr(numeric_only=True)

correlaciones["productivity_score"].sort_values(ascending=False)
```

```python
# GRAFICO
df.plot(
    x="study_hours_per_day",
    y="productivity_score",
    kind="scatter",
    alpha=0.3
)

plt.title("Relación entre horas de estudio y productividad")
plt.show()
```
